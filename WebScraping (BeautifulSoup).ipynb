## **kienthuckhoahoc.org**

**1. Image Captioning**

```python
list_topics = list()

res = requests.get("https://kienthuckhoahoc.org/")
soup = bs(res.content)
title_div = soup.find('div', class_ = 'navbox clearfix')
titles = title_div.find_all('li')

for element in titles:
  title = element.find('a')
  list_topics.append(title['href'])
```

```python
image_caption = dict()
```

```python
def get_image_and_caption(url):
  res = requests.get(url)
  soup = bs(res.content)
  listImage = soup.find_all('div', class_ = 'img-box')
  for div in listImage:
    target = div.find('img')
    if target:
      if target.get('src') and target.get('alt'):
        image_caption["https://kienthuckhoahoc.org" + target.get('src')] = target.get('alt')

def get_article(res):
  soup = bs(res.content) 
  listArticle = soup.find('div', class_ = 'listview')
  if listArticle:
    li_elements = listArticle.find_all('li')
    for li_element in li_elements:
      article_url = li_element.find('a', href=True)
      if article_url != None:
        get_image_and_caption("https://kienthuckhoahoc.org" + str(article_url['href']))

def web_scraping(topic):
  res = requests.get("https://kienthuckhoahoc.org/" + topic)
  get_article(res)
  for page in range(2, 101):
    res = requests.get("https://kienthuckhoahoc.org/" + topic + f"page{page}")
    get_article(res)
```

```python
for topic in list_topics:
  web_scraping(topic)
```
